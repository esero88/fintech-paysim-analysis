## Dataset Overview

The dataset contains transaction-level data including transaction type, amount, sender and receiver accounts, balances, and fraud indicators.

```python
import pandas as pd

df = pd.read_csv("data/PS_20174392719_1491204439457_log.csv", nrows=200000)

df.head()
# Note: The full dataset contains millions of rows, so a subset is used to improve performance during the first stage.
```

## Fraud Distribution

This step checks how many transactions are fraudulent compared to normal ones.

```python
df['isFraud'].value_counts()
```

## Insight

Most transactions are normal, and only a small number are fraud.

This is similar to real life, where fraud is rare.

## Transaction Types

This shows how many transactions exist for each type.

```python
df['type'].value_counts()
```

## Insight

Most transactions are PAYMENT and CASH_OUT.

TRANSFER and DEBIT transactions are less common.

## Fraud by Transaction Type

This shows how fraud is distributed across different transaction types.

```python
df.groupby('type')['isFraud'].sum()
```

## Insight

Fraud is only seen in TRANSFER and CASH_OUT transactions.

Other transaction types do not show any fraud cases.

## Average Transaction Amount

This shows the average transaction amount for each type.

```python
df.groupby('type')['amount'].mean()
```

## Insight

TRANSFER transactions have the highest average amount.

CASH_OUT transactions also have relatively high values compared to others.

## Fraud vs Non-Fraud Amount Comparison

This compares the average transaction amount between fraudulent and normal transactions.

```python
df[df['isFraud'] == 1]['type'].value_counts()
```

## Insight

Fraud cases are almost evenly distributed between CASH_OUT and TRANSFER transactions.

This suggests that both transaction types carry a similar level of fraud risk.

## Fraud Rate by Transaction Type

This shows the percentage of fraudulent transactions for each transaction type.

```python
df.groupby('type')['isFraud'].mean().sort_values(ascending=False)
```

## Insight

TRANSFER and CASH_OUT transactions have the highest fraud rates.

Other transaction types have almost no fraud cases.

## Fraud vs Non-Fraud Amount

This compares the average transaction amount between fraudulent and normal transactions.

```python
df.groupby('isFraud')['amount'].mean()
```

## Insight

Fraudulent transactions tend to have higher amounts compared to normal transactions.

This suggests that fraudsters may focus on high-value transactions.

## Final Insights

- Fraud is rare, representing a very small portion of total transactions.

- Fraud is concentrated in TRANSFER and CASH_OUT transaction types.

- TRANSFER transactions have the highest fraud rate and highest average amount.

- Fraudulent transactions tend to involve significantly higher amounts than normal transactions.

- This suggests that fraud risk is not random, but linked to specific transaction types and high-value operations.
